```python
from IPython.display import Image
from configparser import ConfigParser
import requests

class GiphyAPITranslate(object):
    def __init__(self):
        self.config = ConfigParser()
        self.config.read('config.ini')
        self.config.sections()
        self.api_key = self.config['default']['api-key']
        self.api_url = "http://api.giphy.com/v1/gifs/translate?s={}&api_key={}"
        
    def translate(self, term='JMU'):
        api_url = self.api_url.format(term, self.api_key)
        response = requests.get(api_url).json()
        url = response['data']['images']['original']['url']
        return Image(url)
```
